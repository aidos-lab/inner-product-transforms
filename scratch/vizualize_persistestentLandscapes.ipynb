```python
import matplotlib.pyplot as plt 
import numpy as np 
model = "Encoder"
features = np.load(f"./results/TopologicalMetrics/features_{model}.npy")
labels = np.load(f"./results/TopologicalMetrics/labels_{model}.npy")
```

```python
cls_mean = np.zeros(shape=(3,2,100))

for label, curve in zip(labels,features):
    cls_mean[int(label)] += curve 
```

```python
plt.plot(cls_mean[0][0],c="c")
plt.plot(cls_mean[0][1],c="c")
plt.plot(cls_mean[1][0],c="r")
plt.plot(cls_mean[1][1],c="r")
plt.plot(cls_mean[2][0],c="b")
plt.plot(cls_mean[2][1],c="b")
```
